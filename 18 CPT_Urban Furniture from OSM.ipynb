```python
from osmnx.features import features_from_bbox
import matplotlib.pyplot as plt

# Define bounding box (west, south, east, north) for Cape Town
bbox = (18.307, -34.391, 19.019, -33.468)

# Tags related to urban furniture and public amenities
tags = {
    # Urban Furniture and Public Amenities
    "amenity": [
        "bench", "bicycle_rack", "fountain", "shelter", "playground", "memorial"
    ],
    "highway": ["bus_stop"]  # This includes bus stops, often associated with urban furniture
}

# Correct function usage: only two arguments — bbox and tags
urban_furniture_gdf = features_from_bbox(bbox, tags)

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
urban_furniture_gdf.plot(ax=ax, color="blue", markersize=5, alpha=0.7)
ax.set_title("Urban Furniture in Cape Town (OSM)", fontsize=14)
ax.axis('off')
plt.show()
```

```python
urban_furniture_gdf.to_file("cpt_urbfurn.gpkg", driver="GPKG")
```

```python
from osmnx.features import features_from_bbox
import matplotlib.pyplot as plt
import geopandas as gpd

# Define bounding box (west, south, east, north) for Cape Town
bbox = (18.307, -34.391, 19.019, -33.468)

# Tags related to urban water features
tags = {
    # Water features
    "natural": ["water"],  # Natural water features (rivers, lakes, etc.)
    "waterway": ["river", "canal", "stream", "drain", "ditch"],  # Waterways (rivers, streams, etc.)
    "water": ["reservoir", "wastewater"]  # Reservoirs, wastewater
}

# Correct function usage: only two arguments — bbox and tags
water_features_gdf = features_from_bbox(bbox, tags)

# Save to GeoPackage
water_features_gdf.to_file("urban_water_features_cape_town.gpkg", driver="GPKG")

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
water_features_gdf.plot(ax=ax, color="cyan", markersize=5, alpha=0.7)
ax.set_title("Urban Water Features in Cape Town (OSM)", fontsize=14)
ax.axis('off')
plt.show()
```
